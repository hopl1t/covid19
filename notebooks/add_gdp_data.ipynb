## add_gdp_data notebook

Adds GDP (in billion USD) column to given tableau and hopkins augmented data.

```python
import pandas as pd
```

```python
tableau_conf_data = pd.read_csv('../augmented_datasets/tableau_conf_data.csv')
tableau_death_data = pd.read_csv('../augmented_datasets/tableau_death_data.csv')
```

```python
GDP_data = pd.read_csv('../external_datasets/MDI_GDP_1.csv')
```

```python
# prepping GDP data to have same Country_Region names as Tableau and Hopkins data

GDP_data.rename(columns = {'Country':'Country_Region'}, inplace = True) 
GDP_data = GDP_data.replace({
    'United States': 'US',
    'Iran, Islamic Rep.': 'Iran',
    'Congo, Rep.': 'Congo (Brazzaville)',
    'Congo, Dem. Rep.': 'Congo (Kinshasa)',
    'St. Lucia': 'Saint Lucia',
    'Czech Republic':'Czechia',
    'Kyrgyz Republic': 'Kyrgyzstan',
    'Egypt, Arab Rep.': 'Egypt',
    'Syrian Arab Republic': 'Syria',
#     'Cabo Verde': 'Cape Verde',
    'Brunei Darussalam': 'Brunei',
    'Slovak Republic': 'Slovakia',
    'Korea, Rep.': 'Korea, South',
    'Timor Leste': 'East Timor',
    'St. Vincent and the Grenadines': 'Saint Vincent and the Grenadines',
    })
# GDP_data.tail(20)
```

```python
# print(pd.DataFrame(pd.merge(conf_data, GDP_data, on='Country_Region', how='left')))
```

```python
def add_GDP_data(df, GDP_data):
    df = df.replace({'Cape Verde': 'Cabo Verde',
                    'Timor-Leste': 'East Timor',
                    'Timor Leste': 'East Timor',
                    'Taiwan*': 'Taiwan'
                    })
    merged_df = df.merge(GDP_data, on='Country_Region', how='left')
    
    merged_df['GDP'][merged_df.Country_Region == 'Taiwan']=589.9

    # Martinique data from 2015source: https://en.populationdata.net/countries/martinique/
    merged_df['GDP'][merged_df.Country_Region == 'Martinique']=9.363

    # Holy See (aka Vatican City) GDP not available and mayb enot relevant?
    # Estimated $21 million for 1999, source: https://www.encyclopedia.com/places/spain-portugal-italy-greece-and-balkans/italian-political-geography/vatican-city
    # merged['GDP'][merged.Country_Region == 'Holy See']=

    # source: https://countryeconomy.com/gdp/venezuela
    merged_df['GDP'][merged_df.Country_Region == 'Venezuela']=98.4

    # source: https://tradingeconomics.com/syria/gdp
    merged_df['GDP'][merged_df.Country_Region == 'Syria']=65

    # source: https://countryeconomy.com/gdp/eritrea
    merged_df['GDP'][merged_df.Country_Region == 'Eritrea']= 2

    return merged_df
```

```python
tableau_conf_gdp = add_GDP_data(tableau_conf_data, GDP_data)
tableau_conf_gdp.to_csv('../augmented_datasets/tableau_conf_with_GDP.csv')
```
