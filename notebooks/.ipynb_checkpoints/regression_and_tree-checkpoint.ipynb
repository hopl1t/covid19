```python
import pickle
import random
import math
import warnings
import sklearn
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from datetime import datetime
from datetime import timedelta
%matplotlib inline
# plt.matplotlib.rcParams.update({'font.size': 50})
warnings.filterwarnings('ignore')
plt.rcParams['figure.figsize'] = (12.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'   
plt.rcParams["patch.force_edgecolor"] = False
plt.rc('figure', titlesize=25)
```

```python
PICKLE_PATH = '../augmented_datasets/pickles/hopkins_confirmed_for_regression0904.pkl'
NEW_PICKLE_PATH = '../augmented_datasets/pickles/final_data/hopkins_conf_narrow_0605.pkl'
DEATH_PICKLE_PATH = '../augmented_datasets/pickles/final_data/hopkins_death_narrow_0605.pkl'
OUTPUT_TREE_PATH = '../products/decision_tree/decision_tree_{0}.dot'
MAX_DT_DEPTH = 5
OUTLIER_QUARTILES = (0.02, 0.98) # Min, max quartiles
```

```python
from sklearn.model_selection import KFold
from sklearn import tree, metrics

def train_test_split(X, n_splits=5):
    '''
    Splits rows into training indices and test indices.
    :param X: numpy array of training data, e.g.  np.array([[1, 2], [3, 4], [1, 2], [3, 4]]) - each sample has two features
    :return: Returns indices of rows for train and test for n_splits. e.g. n_splits=2: 
    train_folds = [[0,2,3], [1,2,3]] test_folds = [[1], [0]] 
    '''

    kf = KFold(n_splits=n_splits, random_state=2346, shuffle=True)
    kf.get_n_splits(X)
    
    train_folds, test_folds = [], []
    
    for train_index, test_index in kf.split(X):
        train_folds.append(train_index)
        test_folds.append(test_index)
    
    return train_folds, test_folds


def decision_tree_train(X_train, y_train):
    dt = tree.DecisionTreeRegressor(max_depth=MAX_DT_DEPTH)
    trained_model = dt.fit(X_train, y_train)
    return trained_model

def decision_tree_train_pruned(X_train, y_train, X_test, y_test):
    dt = tree.DecisionTreeRegressor()
    path = dt.cost_complexity_pruning_path(X_train, y_train)
    ccp_alphas, impurities = path.ccp_alphas, path.impurities
    dt_regressors = []
    for ccp_alpha in ccp_alphas:
        dt = tree.DecisionTreeRegressor(ccp_alpha=ccp_alpha)
        dt.fit(X_train, y_train)
        dt_regressors.append(dt)
    dt_regressors = dt_regressors[:-1]
    ccp_alphas = ccp_alphas[:-1]
    
    test_scores = [dt.score(X_test, y_test) for dt in dt_regressors]
    
    index_best_model = np.argmax(test_scores)
    best_model = dt_regressors[index_best_model]
    
    return best_model
#     dt = tree.DecisionTreeRegressor(max_depth=MAX_DT_DEPTH)
#     trained_model = dt.fit(X_train, y_train)
#     return trained_model

def linear_regression_train(X_train, y_train):
    regr = sklearn.linear_model.LinearRegression()
    trained_model = regr.fit(X_train, y_train)
    return trained_model

def remove_outlier(df, range_, columns):
    low = range_[0]
    high = range_[1]
    qnt = df.quantile([low, high])
    data = df.copy()
    for col in list(columns):
        data[col] = \
            data[(data[col] > qnt.loc[low, col]) &\
               (data[col] < qnt.loc[high, col])][col]
    return data.dropna()

def regression(data_, feature_cols, label_cols, outlier_quartiles=(0.02, 0.98), rmse=False, plot_scatter=''):
    arrays = [[1,1,1,2,2,2,3,3,3,4,4,4,5,5,5],\
              ['test','train','naive','test','train','naive','test','train',\
               'naive','test','train','naive','test','train','naive']]
    idx = pd.MultiIndex.from_arrays(arrays, names=('run', 'mse'))
    regr_rslt = pd.DataFrame(index=idx)
    for col in label_cols:
        regr_rslt[col] = np.nan
    summary = pd.DataFrame(
                          {col: np.nan for col in label_cols},
                          index=['avg((naive - test)/naive)', 'avg((naive - train)/naive)']
                          )
    model_fn = linear_regression_train
    metric_fn = metrics.mean_squared_error
    
    if plot_scatter:
        fig, axes = plt.subplots(2, 5, figsize=(25, 10))
    
    for label_col in label_cols:
        data = data_[data_[label_col] > 0].copy()
        data = remove_outlier(data, outlier_quartiles, [label_col, 'avg_interval_tmp'])
        len(data)

        X, y = np.array(data[feature_cols]), np.array(data[[label_col]])
        # Split the data into train, test for n_splits train-test rounds
        try:
            train_folds, test_folds = train_test_split(X, n_splits=5)
        except ValueError:
            continue
        # Train-test the model for each of the n_splits:
        for train_test_round in range(len(train_folds)):
            train_index = train_folds[train_test_round]
            test_index = test_folds[train_test_round]
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]
            regr_trained = model_fn(X_train, y_train)

            # evaluate on test
            y_pred = regr_trained.predict(X_test)
            regr_rslt.loc[train_test_round + 1, 'test'][label_col] = metric_fn(y_test, y_pred, squared = not rmse)
        
            # Compare to train data
            y_pred_train = regr_trained.predict(X_train)
            regr_rslt.loc[train_test_round + 1, 'train'][label_col] = metric_fn(y_train, y_pred_train,\
                                                                                squared = not rmse)

            # Compare to a naive mean-value model:
            y_pred_naive = np.ndarray(y_pred.shape)
            y_pred_naive.fill(y_train.mean())
            regr_rslt.loc[train_test_round + 1, 'naive'][label_col] = metric_fn(y_test, y_pred_naive,\
                                                                                squared = not rmse)
            # Plot computed vs real if specified
            if label_col == plot_scatter:
                _ = axes[0][train_test_round].set_xlabel('Real values', fontsize='large')
                _ = axes[0][train_test_round].set_ylabel('Computed values', fontsize='large')   
                _ = axes[0][train_test_round].set_title('{0} - round {1}, Computed values, n={2}'\
                                                        .format(label_col, train_test_round, len(y_pred)))
                _ = axes[0][train_test_round].scatter(y_test, y_pred)

                _ = axes[1][train_test_round].set_xlabel('Real values', fontsize='large')
                _ = axes[1][train_test_round].set_ylabel('Naive model values', fontsize='large')   
                _ = axes[1][train_test_round].set_title('{0} - round {1}, Naive values, n={2}'\
                                                        .format(label_col, train_test_round, len(y_pred)))
                _ = axes[1][train_test_round].scatter(y_test, y_pred_naive)
                     
    for col in regr_rslt.columns:
        summary.loc['avg((naive - test)/naive)'][col] = \
        ((regr_rslt[col].loc[:,'naive'] - regr_rslt[col].loc[:,'test']) / regr_rslt[col].loc[:,'naive']).mean()
        summary.loc['avg((naive - train)/naive)'][col] = \
        ((regr_rslt[col].loc[:,'naive'] - regr_rslt[col].loc[:,'train']) / regr_rslt[col].loc[:,'naive']).mean()

    return summary, regr_rslt

def dt(data_, feature_cols, label_cols, outlier_quartiles=(0.02, 0.98),\
       rmse=False, plot_scatter='', prune=False):
    arrays = [[1,1,1,2,2,2,3,3,3,4,4,4,5,5,5],\
              ['test','train','naive','test','train','naive','test','train',\
               'naive','test','train','naive','test','train','naive']]
    idx = pd.MultiIndex.from_arrays(arrays, names=('run', 'mse'))
    regr_rslt = pd.DataFrame(index=idx)
    for col in label_cols:
        regr_rslt[col] = np.nan

    summary = pd.DataFrame(
                          {col: np.nan for col in label_cols},
                          index=['avg((naive - test)/naive)', 'avg((naive - train)/naive)']
                          )
    
    if prune:
        model_fn = decision_tree_train_pruned
    else:
        model_fn = decision_tree_train

    metric_fn = metrics.mean_squared_error

    if plot_scatter:
        fig, axes = plt.subplots(2, 5, figsize=(25, 10))
    
    for label_col in label_cols:
        data = data_[data_[label_col] > 0].copy()
        data = remove_outlier(data, outlier_quartiles, [label_col, 'avg_interval_tmp'])

        X, y = np.array(data[feature_cols]), np.array(data[[label_col]])
        # Split the data into train, test for n_splits train-test rounds
        try:
            train_folds, test_folds = train_test_split(X, n_splits=5)
        except ValueError:
            continue
        # Train-test the model for each of the n_splits:
        for train_test_round in range(len(train_folds)):
            train_index = train_folds[train_test_round]
            test_index = test_folds[train_test_round]
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]
            
            if prune:
                dt_trained = model_fn(X_train, y_train, X_test, y_test)
            else:
                dt_trained = model_fn(X_train, y_train)
 
            # Save plot to file, to open: dot -Tpng tree.dot -o tree.png
            suffix = label_col + '_run' + str(train_test_round)
            _ = tree.export_graphviz(dt_trained, OUTPUT_TREE_PATH.format(suffix), feature_names=feature_cols) 
            
            # evaluate on test
            y_pred = dt_trained.predict(X_test)
            regr_rslt.loc[train_test_round + 1, 'test'][label_col] = metric_fn(y_test, y_pred, squared = not rmse)
            
            # Compare to train data
            y_pred_train = dt_trained.predict(X_train)
            regr_rslt.loc[train_test_round + 1, 'train'][label_col] = metric_fn(y_train, y_pred_train,\
                                                                               squared = not rmse)

            # Compare to a naive mean-value model:
            y_pred_naive = np.ndarray(y_pred.shape)
            y_pred_naive.fill(y_train.mean())
            regr_rslt.loc[train_test_round + 1, 'naive'][label_col] = metric_fn(y_test, y_pred_naive,\
                                                                               squared = not rmse)
            # Plot computed vs real if specified
            if label_col == plot_scatter:
                _ = axes[0][train_test_round].set_xlabel('Real values', fontsize='large')
                _ = axes[0][train_test_round].set_ylabel('Computed values', fontsize='large')   
                _ = axes[0][train_test_round].set_title('{0} - round {1}, Computed values, n={2}'\
                                                        .format(label_col, train_test_round, len(y_pred)))
                _ = axes[0][train_test_round].scatter(y_test, y_pred)

                _ = axes[1][train_test_round].set_xlabel('Real values', fontsize='large')
                _ = axes[1][train_test_round].set_ylabel('Naive model values', fontsize='large')   
                _ = axes[1][train_test_round].set_title('{0} - round {1}, Naive values, n={2}'\
                                                        .format(label_col, train_test_round, len(y_pred)))
                _ = axes[1][train_test_round].scatter(y_test, y_pred_naive)
            
    for col in regr_rslt.columns:
        summary.loc['avg((naive - test)/naive)'][col] = \
        ((regr_rslt[col].loc[:,'naive'] - regr_rslt[col].loc[:,'test']) / regr_rslt[col].loc[:,'naive']).mean()
        summary.loc['avg((naive - train)/naive)'][col] = \
        ((regr_rslt[col].loc[:,'naive'] - regr_rslt[col].loc[:,'train']) / regr_rslt[col].loc[:,'naive']).mean()

    return summary, regr_rslt

```

###### Prepare data

Some of the 'first_7' rows are nans (for samples where n>20 but 7 days have not passed)

Some of the Province_State are nans (Not all countries have provinces)

```python
hopkins_confirmed = pd.read_pickle(NEW_PICKLE_PATH)
```

```python
nan_countries = []
nan_first_7 = []
for column in hopkins_confirmed.columns:
    any_nan = hopkins_confirmed[column].loc[:, 'data'].isna().any()
    nulls = hopkins_confirmed[column].loc[:, 'data'].isna().tolist()
    if any_nan:
        size = len([nul for nul in nulls if nul])
        print('{0}: {1}, {2}'.format(column, any_nan, size))
        for i in range(len(nulls)):
            if column == 'Country_Region':
                nc = hopkins_confirmed['Country_Region'].index[i][0]
                nan_countries.append(nc)
            if column == 'first_7':
                nc = hopkins_confirmed['first_7'].index[i][0]
                nan_first_7.append(nc)
```

```python
hopkins_confirmed.drop(columns='Total Tests', inplace=True)
hopkins_confirmed.drop(columns='State Population', inplace=True)
hopkins_confirmed['Democracy'] = hopkins_confirmed['Democracy'].fillna(hopkins_confirmed['Democracy'].mean())
hopkins_confirmed['Tests \\ Pop'] = hopkins_confirmed['Tests \\ Pop'].fillna(0)
hopkins_confirmed['first_7'].dropna(inplace=True)
hopkins_confirmed['GDP'].dropna(inplace=True)
hopkins_confirmed['Urbanization'].dropna(inplace=True)
```

```python
# hopkins_confirmed.drop(nan_first_7, level=0, inplace=True) # This was removed because it was buggy. the dropna is better and robust
hopkins_confirmed['Province_State'] = hopkins_confirmed['Province_State'].fillna('')
hopkins_confirmed.dropna(inplace=True)
# Normalize first_7
hopkins_confirmed['first_7'] = ((hopkins_confirmed['first_7'] + 1) ** (1/7) - 1)
hopkins_confirmed.insert(4, 'tmp**2', np.nan)
hopkins_confirmed.insert(4, 'tmp*rh', np.nan)
hopkins_confirmed.insert(4, 'urb**2', np.nan)
hopkins_confirmed.insert(4, 'tpc*max_cases', np.nan)
hopkins_confirmed.insert(4, 'tests_per_million**2', np.nan)
hopkins_confirmed.insert(4, 'imobility**2', np.nan)

hopkins_confirmed['tmp**2'] = hopkins_confirmed['avg_interval_tmp'] ** 2
hopkins_confirmed['tmp*rh'] = hopkins_confirmed['avg_interval_tmp'] * hopkins_confirmed['avg_interval_RH']
hopkins_confirmed['urb**2'] = hopkins_confirmed['Urbanization'] ** 2
hopkins_confirmed['tpc*max_cases'] = hopkins_confirmed['Tests \ Pop'] * hopkins_confirmed['Max_Cases']
hopkins_confirmed['tests_per_million**2'] = hopkins_confirmed['Tests per 1M'] ** 2
hopkins_confirmed['imobility**2'] = hopkins_confirmed['avg_mobility'] ** 2
```

###### Regression and Tree

```python
us_feature_cols = ['GDP', 'avg_interval_tmp', 'avg_interval_RH',\
                   'tmp**2', 'tmp*rh', 'Urbanization', 'tpc*max_cases', 'Democracy',\
                   'Median Age', 'Tests \ Pop', 'Gini Index', 'imobility**2']

global_feature_cols = ['GDP', 'avg_interval_tmp', 'avg_interval_RH',\
                       'tmp**2', 'tmp*rh', 'Urbanization', 'Democracy', 'Median Age',\
                       'tests_per_million**2', 'imobility**2', 'Gini Index']

label_cols = ['GF_Q1', 'GF_Q3']

coords = list(set(hopkins_confirmed.index.get_level_values(0)))
euro_coords = [coord for coord in coords if (coord[0] > 36.7 and coord[1] > -16 and coord[1] < 25)]
thrsh_data = hopkins_confirmed[hopkins_confirmed['GDP'] >= 40000]\
                        [hopkins_confirmed['Urbanization'] >= 88].copy()


us_regr_sum, us_regr_full = regression(hopkins_confirmed[hopkins_confirmed['Country_Region'] =='US'].copy(),\
                                       us_feature_cols, label_cols, rmse=False)

us_regr_sum_rmse, us_regr_full_rmse = regression(hopkins_confirmed[hopkins_confirmed['Country_Region'] =='US'].copy(),\
                                               us_feature_cols, label_cols, rmse=True)

us_dt_sum, us_dt_full = dt(hopkins_confirmed[hopkins_confirmed['Country_Region'] =='US'].copy(),\
                           us_feature_cols, label_cols, rmse=False, prune=True)


euro_regr_sum, euro_regr_full = regression(hopkins_confirmed.loc[euro_coords, :, :].copy(),\
                                          global_feature_cols, label_cols, rmse=False)

euro_dt_sum, euro_dt_full = dt(hopkins_confirmed.loc[euro_coords, :, :].copy(),\
                              global_feature_cols, label_cols, rmse=False, prune=True)

global_regr_sum, global_regr_full = regression(hopkins_confirmed, global_feature_cols, label_cols, rmse=False)

global_dt_sum, global_dt_full = dt(hopkins_confirmed, global_feature_cols, label_cols, rmse=False, prune=True)

thrsh_regr_sum, thrsh_regr_full = regression(thrsh_data, global_feature_cols, label_cols, rmse=False)

thrsh_dt_sum, thrsh_dt_full = dt(thrsh_data, global_feature_cols, label_cols, rmse=False, prune=True)


print('\nregression - US MSE')
us_regr_sum
print('\nregression - US RMSE')
us_regr_sum_rmse

print('Decision tree - US MSE (pruned)')
us_dt_sum

print('\nregression - Europe MSE')
euro_regr_sum

print('Decision tree - Europe MSE (pruned)')
euro_dt_sum

print('\nregression - global with thresholds MSE')
thrsh_regr_sum

print('Decision tree - global with thresholds MSE (pruned)')
thrsh_dt_sum

print('\nregression - global MSE')
global_regr_sum

print('Decision tree - global MSE (pruned)')
global_dt_sum
```
